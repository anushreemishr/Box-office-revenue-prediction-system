```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cleaned_imdb_dataset.csv')
df.head()
```

```python
features = ['Votes', 'Rating', 'Metascore']
X = df[features].values
y = df['Revenue (Millions)'].values

# Add a column of ones to X for the intercept (c)
X_b = np.c_[np.ones((X.shape[0], 1)), X] 

# 2. Calculate coefficients using the Normal Equation
# beta = (X.T * X)^-1 * X.T * y
beta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# 3. Interpret coefficients
intercept = beta[0]
coefficients = beta[1:]
print(f"Intercept: {intercept}")
for feature, coef in zip(features, coefficients):
    print(f"Coefficient for {feature}: {coef}")

# 4. Make Predictions and Evaluate
y_pred_multi = X_b.dot(beta)

mse_multi = np.mean((y - y_pred_multi)**2)
rmse_multi = np.sqrt(mse_multi)
ss_res = np.sum((y - y_pred_multi)**2)
ss_tot = np.sum((y - np.mean(y))**2)
r2_multi = 1 - (ss_res / ss_tot)

print(f"\nEvaluation Metrics:\nMSE: {mse_multi:.2f}\nRMSE: {rmse_multi:.2f}\nR2 Score: {r2_multi:.4f}")
from mpl_toolkits.mplot3d import Axes3D

plt.figure()
plt.scatter(df['Votes'], y, label="Actual")

# Predicted values
plt.scatter(df['Votes'], y_pred_multi, label="Predicted")

plt.xlabel("Votes")
plt.ylabel("Revenue")
plt.title("Votes vs Revenue")
plt.legend()
plt.show()
```

```python
# 1. Create a polynomial feature (Degree 2 for 'Votes')
df['Votes_sq'] = df['Votes'] ** 2

# 2. Prepare features (Intercept, Votes, Votes_sq)
X_poly = df[['Votes', 'Votes_sq']].values
X_poly_b = np.c_[np.ones((X_poly.shape[0], 1)), X_poly]

# 3. Solve using Normal Equation
beta_poly = np.linalg.inv(X_poly_b.T.dot(X_poly_b)).dot(X_poly_b.T).dot(y)
y_pred_poly = X_poly_b.dot(beta_poly)

# 4. Evaluation
r2_poly = 1 - (np.sum((y - y_pred_poly)**2) / np.sum((y - np.mean(y))**2))
print(f"Polynomial (Degree 2) R2 Score: {r2_poly:.4f}")

# Sort Votes for smooth curve
sorted_idx = np.argsort(df['Votes'].values)

Votes_sorted = df['Votes'].values[sorted_idx]
y_sorted = y[sorted_idx]
y_pred_sorted = y_pred_poly[sorted_idx]

# Plot
plt.figure()

# Actual Data
plt.scatter(Votes_sorted, y_sorted, label="Actual Data")

# Polynomial Curve
plt.plot(Votes_sorted, y_pred_sorted,color="red", label="Polynomial Fit (Degree 2)", linewidth=2)

plt.xlabel("Votes")
plt.ylabel("Revenue (Millions)")
plt.title("Polynomial Regression (Degree 2)")
plt.legend()

plt.show()
```

```python
# 1. Feature Scaling (Highly recommended for Regularization)
X_scaled = (X - X.mean(axis=0)) / X.std(axis=0)
X_scaled_b = np.c_[np.ones((X_scaled.shape[0], 1)), X_scaled]

# 2. Ridge Regression Implementation
alpha = 1.0  # Regularization strength (lambda)
I = np.identity(X_scaled_b.shape[1])
I[0, 0] = 0  # Don't regularize the intercept

# beta = (X.T * X + alpha * I)^-1 * X.T * y
beta_ridge = np.linalg.inv(X_scaled_b.T.dot(X_scaled_b) + alpha * I).dot(X_scaled_b.T).dot(y)

print("Ridge Coefficients:", beta_ridge[1:])
```

```python
# Plot Residuals vs Predicted Values
residuals = y - y_pred_multi

plt.scatter(y_pred_multi, residuals, alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Predicted Revenue")
plt.ylabel("Residuals (Errors)")
plt.title("Residual Plot (Model Diagnostics)")
plt.show()

print("Observation: If residuals are randomly scattered around the zero-line, the model is valid.")
```
